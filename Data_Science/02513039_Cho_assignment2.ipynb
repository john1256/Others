```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1:
 Loading the 'titanic' dataset from the online repository provided by the
seaborn library
â€¢ Requires internet

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

# Problem 1: 
Visualize the number of survivor by gender

```python
sns.countplot(x = 'sex', data = titanic, hue='survived')
plt.title('Survival by Gender')
```

# Problem 2: 
Visualize the number of survivor by passenger class

```python
sns.countplot(x = 'pclass', data = titanic, hue='survived')
plt.title('Survival by PClass')
```

# Problem 3:
 Visualize the number of people per passenger class by embarked port

```python
sns.countplot(x='embarked', data = titanic, hue='pclass')
plt.title('PClass by Embarked')
```

# Problem 4: 
Visualize survival rate by gender and passenger class

```python
sns.barplot(x=  'sex', y='survived', data=titanic, hue='pclass')
plt.title('Survival Rate by gender and PClass')
```

# Problem 5:
 Visualize age distribution by embarked port and gender.

```python
sns.violinplot(x = 'embarked', y = 'age', data = titanic, hue='sex', split=True)
```

# Problem 6:
 Visualize the survival by gender and passenger class

```python
sns.catplot(x ='sex', data = titanic, kind='count', hue='survived', col='pclass')
```
