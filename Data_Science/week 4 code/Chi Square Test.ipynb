```python
import pandas as pd
data=pd.read_csv('ex7-4.csv')
data.head()
```

```python
data.describe()
```

```python
#construct frequency table 
pd.crosstab(index=data['amount'], columns=data['level'], values=data['count'], aggfunc=sum,\
            margins=True, margins_name='total')
```

```python
#construct probability table
pd.crosstab(index=data['amount'], columns=data['level'], values=data['count'], aggfunc=sum, \
            margins=True, margins_name='total', normalize='index').round(4)
```

```python
#Chi-squared test and  constructing Expected Frequency Table
from scipy.stats import chi2_contingency
d_table=pd.crosstab(index=data['amount'], columns=data['level'], values=data['count'], aggfunc=sum, \
                    margins=True, margins_name='전체')

chi,p,df,expected=chi2_contingency(d_table)

expected
```

```python

expected_table=pd.DataFrame(data=expected, index=d_table.index, columns=d_table.columns)
expected_table

print(chi, p)
```
