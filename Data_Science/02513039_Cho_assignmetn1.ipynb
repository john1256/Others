```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

# Problem 1:
 Draw the distribution according to 'Age' as a histogram.
• Bins can be set freely.

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=20, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.title('Distribution of Age')
plt.show()
```

# Problem 2: 
Visualize the ratio of survivors(“Survived==1”) and non-survivors (“Survived==0”)
using the pie chart

```python
survived= df['Survived'].value_counts()
labels = ['Not Survived', 'Survived']
plt.figure(figsize=(6, 6))
plt.pie(survived, labels=labels, startangle=90, autopct='%1.1f%%')
plt.title('Survived passengers')
plt.axis('equal')
plt.show()
```

# Problem 3: 
Visualize the survival rate by gender (‘Sex’) using the bar chart
• hint) using the groupby()

```python
survived = df.groupby('Sex')['Survived'].mean()
plt.figure(figsize=(7, 6))
survived.plot(kind='bar', color=['red', 'blue'])
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
```

# Problem 4: 
Visualize the survival rate by passenger class (‘Pclass’) using the bar chart
• hint) using the groupby()

```python
survived = df.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(7, 6))
survived.plot(kind='bar', color=['green', 'orange', 'blue'])
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
```

# Problem 5: 
Visualize the survival rate by fare category using the bar chart
• Divide ‘Fare’ into four categories (bins).
• hint) using the cut() and groupby()

```python
fare = pd.cut(df['Fare'], bins=[0,25,50,100,df['Fare'].max()], labels=['0-25', '25-50', '50-100', '100+'])
survived = df.groupby(fare)['Survived'].mean()
plt.figure(figsize=(7, 6))
survived.plot(kind='bar')
plt.xlabel('Fare Range')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Fare Category')
```

# Problem 6:
 Visualize the survival rate by embarked port (‘Embarked’) using the bar chart
• hint) using the groupby()

```python
survived = df.groupby('Embarked')['Survived'].mean()
plt.figure(figsize=(7, 6))
survived.plot(kind='bar')
plt.xlabel('Port')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Port')
plt.show()
```
